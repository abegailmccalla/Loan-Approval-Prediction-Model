```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv('/kaggle/input/comp-3220-loan-prediction/train.csv')
test = pd.read_csv('/kaggle/input/comp-3220-loan-prediction/test.csv')



x = pd.get_dummies(data, columns=['person_gender','person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file' ], drop_first=True)
x_test = pd.get_dummies(test, columns=['person_gender','person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file' ], drop_first=True)


x = x.drop(columns=['loan_status', 'loan_id']) 
x_test = x_test.drop(columns=['loan_id']) 


x = x.astype(int)

x_test.head()
```

```python
sc = StandardScaler()

x1 = x.values
x1_test = x_test.values

y = data['loan_status'].values

X = sc.fit_transform(x1)
X_test = sc.fit_transform(x1_test)

print(X_test)
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(X, y)
```

```python
predictions = clf.predict(X_test)

results = pd.DataFrame({
    'loan_id': test['loan_id'],
    'loan_status': predictions
})

results.to_csv("/kaggle/working/predict_.csv", index=False)
print("Predictions saved to predict_.csv")

print(predictions[:1000])
```
