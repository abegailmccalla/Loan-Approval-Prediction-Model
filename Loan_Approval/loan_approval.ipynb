```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Limit to the first 1000 rows
data = data.iloc[:1000]

# x_test = test.drop(test.columns[[1, 2, 5, 7, 12, 13]], axis=1)



x = pd.get_dummies(data, columns=['person_gender','person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file' ], drop_first=True)
x_test = pd.get_dummies(test, columns=['person_gender','person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file' ], drop_first=True)


x = x.drop(columns=['loan_status', 'loan_id']) 
x_test = x_test.drop(columns=['loan_id']) 


x = x.astype(int)

x_test.head()
```

```python
x.head()
```

```python
sc = StandardScaler()

x1 = x.values
x1_test = x_test.values

y = data['loan_status'].values

X = sc.fit_transform(x1)
X_test = sc.fit_transform(x1_test)

print(X_test)
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(X, y)
```
